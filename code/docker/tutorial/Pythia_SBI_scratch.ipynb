````python
def generate_thrust_dist(N_events, 
                         proton_electron_beam = 'electron', 
                         parallel=False,
                        which_particles=1):
    """
    which_particles is the argument for Thrust, that tells which particles are analyzed
        1: all final state particles
        2: all observable final state particles, excluding neutrinos and other invisible particles
        3: only charged final state particles
    """
    if parallel:
        pythia = pythia8.PythiaParallel()
        pythia.readString("Parallelism:numThreads = 6")
        # Require each thread to produce the same number of events.
        pythia.readString("Parallelism:balanceLoad = on")
    else:
        pythia = pythia8.Pythia()
        
    if proton_electron_beam == 'electron':
        # Start from the Monash tune.
        pythia.readString("Tune:ee = 7")
        # Set alpha_s to the Monash value.
        # This value is already being used, but this is shown for clarity.
        pythia.readString("TimeShower:alphaSvalue = 0.1365")
        # Set the beams for the LEP configuration.
        pythia.readString("Beams:idA = 11")     # First beam is the electron.
        pythia.readString("Beams:idB = -11")    # Second beam is the positron.
        pythia.readString("Beams:eCM = 91.189") # COM is mass of the Z.
        pythia.readString("PDF:lepton = off")   # Turn off PDFs for lepton beams.
        # Set the process (don't include b-production).
        pythia.readString("WeakSingleBoson:ffbar2gmZ = on")
        pythia.readString("23:onMode = off")
        pythia.readString("23:onIfAny = 1 2 3 4")
    
    
    else:
        # Initialize, incoming pp beams are default.
        pythia.readString("Top:gg2ttbar = on")  # Switch on process.
        pythia.readString("Beams:eCM = 14000.") # Set 14 TeV CM energy.
        pythia.readString("MultipartonInteractions:pT0Ref=%s" % 2.1006)
        
        
    pythia.readString("Main:numberOfEvents = %s" % N_events) # set the number of events to generate.
    
    Thrust_dist=np.empty(N_events)
    Weight_dist=np.empty(N_events)
    pythia.init()
    for iEvent in range(N_events):
        pythia.next()
        weight = pythia.infoPython().Weight()
        Weight_dist[iEvent] = weight
        if pythia8.Thrust(1).analyze(pythia.event):
            THRUST = pythia8.Thrust(1).thrust()
        
            Thrust_dist[iEvent] = THRUST

    return Thrust_dist, Weight_dist
````

````python
THRUST_dist, Weight_dist = generate_thrust_dist(N_events=10000)
````
